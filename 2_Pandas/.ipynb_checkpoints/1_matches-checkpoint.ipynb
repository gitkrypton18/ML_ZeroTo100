```python

def get_city_match_count(city):
    mask=data["city"]==city
    return data[mask].shape[0]
    
```

```python
get_city_match_count('Bangalore')
```

```python
get_city_match_count('Rajkot')
```

### value_counts
: frequency count of the data in one column
: applicable on categorical data only e.g. tem1 column
: The value_counts() function in Pandas is used to get the frequency count of unique values in a Series (i.e., a single column of a DataFrame).
It is especially useful for analyzing categorical data, such as names, cities, or grades.

```python
# har team ne aaj tak kitne match jite hai

data['winner'].value_counts()
```

### Drop Duplicates
built in function to remove duplicates
checks row by row it checks same occurences 


```python
import numpy as np
import pandas as pd
data=pd.read_csv('matches.csv') # This will read the data from the csv value
data
```

```python
data.drop_duplicates(subset=["city"])
```

```python
data.drop_duplicates(subset=["city"]).shape

# to make it permanent use inplace = True
# what 31 is saying means ipl aajtak 31 different cities me khela gaya hai
```

```python
data.drop_duplicates(subset=["city","season"])
```

```python
data.drop_duplicates(subset=["city","season"]).shape

# jaaha jaha pe bhi city aur seasson reoccur kar rhe h 

# and operation jaha jaha city aur season hai wo gayab ho gaya

```

```python
# we want season and winner now har season ka winner 10 season winner

# final match is the last match of the season

data.drop_duplicates('season')
# but it will keep forst occurence

```

```python
data.drop_duplicates('season',keep="last")[["season","winner"]]
# this will display the winnning teams in that particular season
```
