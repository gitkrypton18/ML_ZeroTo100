### Merge Function


```python
import numpy as np
import pandas as pd

match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
# Now I want the list of Ornage Cap Holders in each season
# merge functionn joining the two data sets
# we have a common column id and match_id in both data sets

new=delivery.merge(match,left_on='match_id',right_on='id')
new
```

```python
delivery.merge(match,left_on='match_id',right_on='id').shape
# both are merged
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum()
```

```python
# har season me top batsman
# reset aand set index
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batsman']]
# reset index is used 
# now drop dupliccates and keep only top batsman
```

### Pivot Table

```python
# I will create a pivot table columns = team name, rows=1 to 20 , har team ne kisi particular team ne kitne chake mare
# we just want sixes
mask=delivery['batsman_runs']==6
# This will filter all which have runs as six
six=delivery[mask]
six.shape
```

```python
#using pivot table
six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
```

```python
# making a heat map

import matplotlib.pyplot as plt
import seaborn as sns

```

```python
pt=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
```

```python
sns.heatmap(pt)
```
