### Handling Missing Values


### 1. Dropping missing data using dropna()

```python
data.shape
```

```python
data.dropna()
# axis: 'Axis' = 0,1 for column
   # how: 'AnyAll | lib.NoDefault' = <no_default>,
# kisi bhi row mw ek bhi null value mili so delete kr dia
# if how=All sare columns ke liye value missing
# by default it cleears that whuch have any of the missing values in rows axis = 0 , axis=1
```

```python

data.dropna().shape
```

```python
data.dropna(how='all')
```

```python
data.dropna(axis=1)
```

```python
data.dropna(axis=1,how='all')
```

```python
data.dropna(subset=['Cabin']).shape
# out off 891 only 204 has the value of cabin
# subset ek ya multipple columns
```

```python
data.dropna(subset=['Cabin','Embarked']).shape
```
